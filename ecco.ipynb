```python
import ecco

# Load pre-trained language model. Setting 'activations' to True tells Ecco to capture neuron activations.
# distillgpt is a distilled GPT2 model. You can also try 'gpt2' 
lm = ecco.from_pretrained("t5-base", activations=True)
lm = ecco.from_pretrained("t5-base", verbose=False)
```

```python
text = "Hi, Where are you going?"
output_2 = lm.generate(text, generate=1, do_sample=False, attribution=['ig'])
```

```python
output_2.rankings()
```

```python
output_2.primary_attributions(attr_method='ig')
```
